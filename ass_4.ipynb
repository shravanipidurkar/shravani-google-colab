# *Drop rows of nan*

```python
import numpy as numpy
import pandas as pd
# Load a CSV file
data = pd.read_csv('/content/all_data.csv')
```

```python
data
```

```python
# find NAN isna() is same as isnull but does not provides the nulled values
nan_df = data[data.isnull().any(axis=1)]
display(nan_df.head)

data= data.dropna(how="all")
data.head()
```

### to work on date column(to handle wrong format)

```python
data = data[data['Order Date'].str[0:2]!='Or']
```

### Make column correct type

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
data
```

### Augment data with additional column (add month data)

```python
data['Month']= data['Order Date'].str[0:2]
#data['Month']= data['Month'].astype('int64')
data.head()
```

```python
#alternative method to add month
import datetime as dt
data['Month 2']= pd.to_datetime(data['Order Date']).dt.month
data.head()
```

### Add city column

```python
def get_city(address):
  return address.split(",")[1].strip(" ")

def get_state(address):
  return address.split(",")[2].strip(" ")[1]

data['City'] = data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
data.head()

```

### Data  Exploration
### Question 1:

```python
data['Sales'] = data['Quantity Ordered'].astype('int') * data['Price Each'].astype('float')5
```
